```
import re
import csv
itemgetter = lambda x: (lambda arr: arr[x])

frequency = {}
with open('wiki.txt', 'r') as wiki:
  text = wiki.read()
#words = text.split()
words = [word.lower() for word in re.findall(r'(\b[A-Za-z][a-z]{0,30}\b)', text)]
 
for word in words:
  frequency[word] = frequency.get(word,0) + 1

with open('slownik.csv', 'w') as csvfile:
  writer = csv.writer(csvfile, delimiter=',')
  for key, value in sorted(frequency.items(), key = itemgetter(1), reverse = True):
      print('{}, {}'.format(key, value))
      
      
x = 1       
with open('slownik.csv', 'w') as csvfile:
  writer = csv.writer(csvfile, delimiter=',')
  for key, value in sorted(frequency.items(), key = itemgetter(1), reverse = True):
    writer.writerow([x, key, value])
    x = x + 1
```

```
import csv
data = []

with open('slownik.csv', 'r') as csvfile:
     reader = csv.reader(csvfile, delimiter=',')
     for row in reader:
        #print (', '.join(row))
        data.append(row)

print(data)
```

```
ids =[]
words = []
counts = []
for row in data:
  ids.append(row[0])
  words.append(row[1])
  counts.append(row[2])
  
ids2 = [int(row[0]) for row in data]  
counts2 = [int(row[2]) for row in data]  

print(ids)
print(ids2)
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
plt.plot(ids2[:500], counts2[:500], 'ob', ms=5, linestyle = 'solid')
plt.show()
```

```
maks = 19763
maks1 = [int(maks/row) for row in ids2]
maks2 = [int(maks/row**2) for row in ids2]
maks3 = [int(maks/row**0.5) for row in ids2]
plt.figure(figsize=(20,15))
plt.plot(ids2[:100], counts2[:100], 'ob', ids2[:100], maks1[:100], 'og', ids2[:100], maks2[:100], 'or', ids2[:100], maks3[:100], 'oy', ms = 5, linestyle = 'solid')
plt.gca().legend(('words','maks/x','maks/x**2','maks/x**0.5'))
plt.show()

```
